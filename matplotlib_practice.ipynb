```python
# Import dependencies (line chart)
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the line plot
plt.plot(x_axis,y_axis)
```

```python
# Create the line plot with ax.plt() (object-oriented interface)
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt() (object-oriented interface alternative)
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() (to create a quick graph using object-oriented method)
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot. (use the show function) **plt.show() should only be used once per session**
plt.plot(x_axis, y_axis)
#plt.show()
```

```python
# Create a plot and and add a label for the legend (using Matplab)
plt.plot(x_axis,y_axis, label='Boston')

# Create lables for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare ($)')

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title('PyBer Fare by Month')

# Add the legend
plt.legend()
```

```python
## Create a plot, add a label for the legend, and set parameters for line color, width, and marker (using matlab)
plt.plot(x_axis,y_axis, label='Boston', marker='*', color='blue', linewidth=2)

# Create lables for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare ($)')

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title('PyBer Fare by Month')

## Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Create and annotate plot with object-oriented approach
ax = plt.axes()
ax.plot(x_axis, y_axis, label='Boston', marker='*', color='green', linewidth=2)

# 4. add a legend for the city of boston
ax.legend()

# 5. add a title and axis labels\
ax.set_title('PyBear Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')

#6. Add a y-axis limit
ax.set_ylim(0,45)

# 7. add grid lines
ax.grid()
```

```python
# Bar graphs
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a vertical bar chart (Matplab)
plt.bar(x_axis,y_axis)
```

```python
# Create and annotate bar chart (matplot)
plt.bar(x_axis,y_axis, color="green", label='Boston')

# Create labels for x and y axes
plt.xlabel('Date')
plt.ylabel('Fare ($)')

# Create a title
plt.title('PyBer Fare by Month')

# Add a legend
plt.legend()
```

```python
# Create a HORIZONTAL bar chart (matplot)
plt.barh(x_axis,y_axis)
```

```python
# Switch axes of horizontal bar chart (matplot)
plt.barh(y_axis, x_axis)
```

```python
# Invert y-axis to have Jan-Dec in order - horizontal bar (matlab)
plt.barh(x_axis,y_axis)
plt.gca().invert_yaxis()
```

```python
# Invert y-axis to have Jan-Dec in order - horizontal bar (matlab)
plt.barh(x_axis,y_axis, color='magenta', label='Boston')
plt.gca().invert_yaxis()

#add titles
plt.title('PyBer by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')

plt.legend()
```

```python
# Create bar charts using Object-Oriented Approach
fig, ax=plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
# Create a HORIZONTAL bar chart (object-oriented)
fig, ax=plt.subplots()
ax.barh(x_axis,y_axis)
```

```python
# Switch axes of horizontal bar chart (object-oriented)
fig, ax=plt.subplots()
ax.barh(x_axis,y_axis, color='cyan', label='Chicago')
ax.invert_yaxis()

ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.legend()
```

```python
# Create scatter plot (matlab)
plt.plot(x_axis,y_axis, 'o')
```

```python
# Create scatter plot (matlab)
plt.scatter(x_axis, y_axis)
```

```python
# Create scatter plot (matlab) - skill drill
plt.scatter(y_axis,x_axis, c='red', label='Chicago')

plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')

plt.xlim(0,50)

plt.gca().invert_yaxis()

plt.legend()
```

```python
# Create scatter plot (Object-Oriented Interface)
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
# Create a bubble chart (matlab) - change size
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# adjust size of bubbles

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis,y_axis, s=y_axis_larger)
```

```python
# Create the scatter using list comprehension techniques inside the function
# format for list comprehension is: new_list = [expression for item in lsit if conditional]

plt.scatter(x_axis,y_axis, s = [i*3 for i in y_axis])
```

```python
# Create a bubble chart (Object-Oriented Interface)
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis, s=y_axis)
```

```python
# Create a bubble chart (Object-Oriented Interface) - skill drill
fig, ax=plt.subplots()
ax.scatter(y_axis,x_axis, s = [i*5 for i in y_axis], c='skyblue', label='Boston', alpha=0.8, edgecolors ='black', linewidth=2)

ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')

ax.invert_yaxis()

ax.set_xlim(0,50)

ax.legend()
```
